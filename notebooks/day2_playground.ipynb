```python
import os
from pathlib import Path

from dotenv import load_dotenv
from openai import OpenAI
```

```python
# Load variables from .env into environment variables.
load_dotenv()
```

```python
# client.py

def generate_notes(text: str) -> str:
    """
    Send raw text to the LLM and return the model's raw text output.

    This is intentionally simple for Day 2:
    - We only want to prove the pipeline works.
    - We are NOT enforcing structured JSON yet.
    - We return raw text so we can inspect model behavior first.
    """

    # Read API settings from environment variables.
    # Do not hardcode secrets in source code.
    api_key = os.getenv("OPENAI_API_KEY")
    model = os.getenv("OPENAI_MODEL", "gpt-5.4-mini")

    if not api_key:
        raise ValueError(
            "OPENAI_API_KEY is not set. Please export it before running the script."
        )

    # Initialize the OpenAI client.
    client = OpenAI(api_key=api_key)

    # Keep the prompt simple on Day 2.
    # The goal is to get a useful first output, not perfect structure.
    prompt = f"""
                You are a helpful assistant for note-taking.

                Read the following long-form text and produce:
                1. A short summary
                2. 3 to 5 key points
                3. 2 to 4 action items
                4. 3 to 5 tags

                Text:
                {text}
                """.strip()

    # Use the Responses API to generate text output.
    response = client.responses.create(
        model=model,
        input=prompt,
    )

    # output_text gives us the combined text response directly.
    return response.output_text.strip()
```

```python
# main.py

def load_input_text(file_path: Path) -> str:
    """
    Read the input text file and return its content as a string.
    """
    if not file_path.exists():
        raise FileNotFoundError(f"Input file not found: {file_path}")

    return file_path.read_text(encoding="utf-8").strip()


def main() -> None:
    """
    Minimal Day 2 pipeline:
    1. Load sample input text
    2. Send it to the LLM
    3. Print the raw output
    """
    input_file = Path("../examples/input_chinese.txt")
    text = load_input_text(input_file)

    print("=== Input Loaded ===")
    print(f"Characters: {len(text)}")
    print()

    result = generate_notes(text)

    print("=== Model Output ===")
    print(result)
```

```python
main()
```
